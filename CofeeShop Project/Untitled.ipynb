```python
#range of transaction
date_range=data['transaction_date'].agg([min, max])
```

```python
#number of transaction
num_transactions = data.shape[0]
```

```python
# Get unique products and product categories
unique_products = data['product_type'].unique()
unique_categories = data['product_category'].unique()
```

**Add a Column to Calculate Revenue**

```python
#revenue calculation
data['revenue']=data['transaction_qty']* data['unit_price']
```

```python
data.head(10)
```

```python
data['revenue'].dtype
```

**Add Columns for Month and Day of Week**

```python
#Get month and day of the week from the transaction_date
data['month']= data['transaction_date'].dt.strftime('%b')
data['month_number']= data['transaction_date'].dt.month
data['week'] = data['transaction_date'].dt.strftime('%a')
```

```python
data.head(10)
```

**Add a Column to Extract Hour from Transaction Time**

```python
# Extract the hour directly from the time object
data['hour'] = data['transaction_time'].apply(lambda x: x.hour)
```
